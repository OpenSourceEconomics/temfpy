# Multiple starting values

This notebook showcases the usefulness of `temfpy` to investigate the robustness of optimization results to different starting values. We import a suitable test function from the library. For details on the `ackley` function itself, plese consult our online documentation at [temfpy.readthedocs.org](https://temfpy.readthedocs.io/en/latest/).

```python
from temfpy.optimization import ackley
```

We will leave the parameterization of the function set to its default values $A = 10$ and also focus on optimization problems of dimension $n = 5$. In what follows we simply draw 15 alternative starting values and run the Powell's algorithm as implemented in `scipy`.

```python
from scipy.optimize import minimize
import numpy as np

np.random.seed(1234)

rslt = list()
for iter_ in range(15):
    x0 = np.random.uniform(low=-5.12, high=5.12, size=5)
    fun = minimize(ackley, x0, method="Nelder-Mead")["fun"]

    rslt.append(fun)
```

Now we are able to visualize the results.

```python
import matplotlib.pyplot as plt

ax = plt.subplot()

ax.plot(rslt)

ax.set_ylabel("Function value")
ax.set_xlabel("Random point")

plt.show()
```
